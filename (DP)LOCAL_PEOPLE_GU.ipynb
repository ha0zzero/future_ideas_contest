```python
import pandas as pd
```

##### 2019

```python
p2019= pd.read_csv('./data/LOCAL_PEOPLE_GU_2019.csv', encoding='CP949')
p2019.reset_index(inplace=True)
p2019 = p2019.shift(1, axis = 1)
p2019 = p2019.iloc[:, 1:5]
p2019.rename(columns = {'총생활인구수':'내국인총생활인구수'},inplace=True)
p2019
```

```python
f2019 = pd.read_csv('./data/LONG_FOREIGNER_GU_2019.csv', encoding='CP949')
f2019 = f2019.iloc[:, :4]
f2019.drop(['시간대구분'], axis=1, inplace=True)
f2019.rename(columns = {'총생활인구수':'장기외국인총생활인구수'},inplace=True)
f2019
```

```python
t2019 = pd.read_csv('./data/TEMP_FOREIGNER_GU_2019.csv', encoding='CP949')
t2019 = t2019.iloc[:, :4]
t2019.rename(columns = {'stdr_de_id':'기준일ID', 'adstrd_code_se':'자치구코드', 'tot_lvpop_co':'단기외국인총생활인구수'},inplace=True)
t2019.drop(['tmzon_pd_se'], axis=1, inplace=True)
t2019
```

```python
p2019 = p2019.groupby('자치구코드').mean()
p2019 = p2019.iloc[:, 2:3]
p2019.rename(columns = {'내국인총생활인구수':'2019내국평균'},inplace=True)
p2019
```

```python
f2019 = f2019.groupby('자치구코드').mean()
f2019 = f2019.iloc[:, 1:3]
f2019.rename(columns = {'장기외국인총생활인구수':'2019장기평균'},inplace=True)
f2019
```

```python
t2019 = t2019.groupby('자치구코드').mean()
t2019 = t2019.iloc[:, 1:3]
t2019.rename(columns = {'단기외국인총생활인구수':'2019단기평균'},inplace=True)
t2019
```

```python
df19 = pd.merge(p2019,f2019, how='inner',on='자치구코드')
df19 = pd.merge(df19, t2019, how='inner', on='자치구코드')
df19[2019] = (df19['2019내국평균']+df19['2019장기평균']+df19['2019단기평균'])
```

```python
df19.drop(['2019내국평균'], axis=1, inplace=True)
df19.drop(['2019장기평균'], axis=1, inplace=True)
df19.drop(['2019단기평균'], axis=1, inplace=True)
df19
```
