```python
import pandas as pd
```

```python
p2019= pd.read_csv('./data/LOCAL_PEOPLE_GU_2019.csv', encoding='CP949')
p2019.reset_index(inplace=True)
p2019 = p2019.shift(1, axis = 1)
p2019 = p2019.iloc[:, 1:5]
p2019
```

```python
p2019 = p2019.groupby('자치구코드').mean()
p2019 = p2019.iloc[:, 2:3]
p2019.rename(columns = {'총생활인구수':'2019'},inplace=True)
p2019
```

```python
p2020= pd.read_csv('./data/LOCAL_PEOPLE_GU_2020.csv', encoding='CP949')
p2020 = p2020.iloc[:, 1:4]
p2020 = p2020.groupby('자치구코드').mean()
p2020.drop(['시간대구분'], axis=1, inplace=True)
p2020.rename(columns = {'총생활인구수':'2020'},inplace=True)
p2020
```

```python
p2021= pd.read_csv('./data/LOCAL_PEOPLE_GU_2021.csv', encoding='CP949')
p2021 = p2021.iloc[:, 1:4]
p2021 = p2021.groupby('자치구코드').mean()
p2021.drop(['시간대구분'], axis=1, inplace=True)
p2021.rename(columns = {'총생활인구수':'2021'},inplace=True)
p2021
```

```python
p2022= pd.read_csv('./data/LOCAL_PEOPLE_GU_2022.csv', encoding='CP949')
p2022 = p2022.iloc[:, 1:4]
p2022 = p2022.groupby('자치구코드').mean()
p2022.drop(['시간대구분'], axis=1, inplace=True)
p2022.rename(columns = {'총생활인구수':'2022'},inplace=True)
p2022
```

```python
p2023= pd.read_csv('./data/LOCAL_PEOPLE_GU_2023.csv', encoding='CP949')
p2023 = p2023.iloc[:, 1:4]
p2023 = p2023.groupby('자치구코드').mean()
p2023.drop(['시간대구분'], axis=1, inplace=True)
p2023.rename(columns = {'총생활인구수':'2023'},inplace=True)
p2023
```

```python
p1 = pd.merge(p2019,p2020, how='inner',on='자치구코드')
p2 = pd.merge(p1, p2021, how='inner', on='자치구코드')
p3 = pd.merge(p2, p2022, how='inner', on='자치구코드')
LPG = pd.merge(p3, p2023, how='inner', on='자치구코드')
```

```python
LPG
```

```python
LPG['5개년평균']=(LPG['2019']+LPG['2020']+LPG['2021']+LPG['2022']+LPG['2023'])/5
LPG
```
