```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df= pd.read_excel('./merge_data.xlsx')
df
```

```python
df.columns
```

##### 변수 설명
- 'Seoul_GU' : 서울시 자치구
- 'Metro_count' : 자치구 내 지하철 역 개수
- 'Drinking_fountain' : 자치구 내 음수대 개수
- 'Bus_count' : 자치구 내 버스 정류장 개수
- 'settled_pp' : 자치구 5개년 평균 상주 인구 수
- 'local_pp': 자치구 5개년 평균 생활 인구 수
- 'Public_sports_count' : 자치구 내 공공 체육시설 개수
- 'single_residential_area' : 자치구 내 단독 주거지역 면적
- 'communal_residential_area': 자치구 내 공동 주거지역 면적
- 'administrative_area' : 자치구 내 행정 시설 면적
- 'commercial_area' : 자치구 내 상업 지역 면적
- 'residential_commercial_mix_area' : 자치구 내 주거 및 상업 혼합지역 면적
- 'cultural_recreation_area' : 자치구 내 휴양지역(공원 등) 면적
- 'area(km^2)' : 자치구 면적 합

##### 상관계수 계산

```python
df1 = df.iloc[:,1:]
correlation_matrix = df1.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

###### [결과] 
- 0.5 이상의 상관관계를 가지는 애들
    - 휴양지역 > 상주인구 > 생활인구
