```python
import pandas as pd
```

```python
df= pd.read_csv('./data/서울시 상권분석서비스(상주인구-자치구).csv', encoding='CP949')
df = df.iloc[:, :4]
df = df.sort_values(by = ['자치구_코드', '기준_년분기_코드'])
df.rename(columns = {'자치구_코드':'자치구코드'},inplace=True)
df = df.groupby('자치구코드').mean()
df
```

```python
df.drop(['기준_년분기_코드'], axis=1, inplace=True)
df
```

```python
df.rename(columns = {'총_상주인구_수':'5개년평균'},inplace=True)
```

```python
info= pd.read_excel('./data/행정동코드_매핑정보_20200325.xlsx', sheet_name="유입지코드")
info = info.iloc[0:25,0:3]
info.drop(['RESD_DO_NM'], axis=1, inplace=True)
info.rename(columns = {'RESD_CD':'자치구코드', 'RESC_CT_NM':'자치구'},inplace=True)
info
```

```python
info = info.set_index("자치구코드")
df = pd.merge(info, df, how='inner', on='자치구코드')
df
```

```python
df.to_csv("SETTLED_PEOPLE_GU.csv", encoding='CP949')
```
